
### 3. Are the differences between types of species and their conservation status significant?



```python
# For significance, look at chi-squared. Summarize 'of_concern' and 'non-concerned' for each category. 

species_all['of_concern'] = species_all.conservation_status != 'Not Endangered'

status_concern_vs_not = species_all.groupby(['category', 'of_concern'])\
                .scientific_name.nunique()\
                .reset_index()\
                .pivot(columns='of_concern', index='category',values='scientific_name')\
                .reset_index()
status_concern_vs_not.columns = ['category', 'non_concerned', 'of_concern']
print(status_concern_vs_not)
```

```python
# Write function to take data from table above and run chi2 contingency

def assign_tag(index):
    lst = []
    lst.append(status_concern_vs_not.loc[index, 'of_concern'])
    lst.append(status_concern_vs_not.loc[index, 'non_concerned'])
    return lst      

def run_chi2(index1, index2):
    test_lst = []
    test_lst.append(assign_tag(index1))
    test_lst.append(assign_tag(index2))
    print(f'Results of chi2 contingency comparing {index1} and {index2}')
    print(chi2_contingency(test_lst))
    val = round((chi2_contingency(test_lst)[1]), 3)
    sig = 'Significant' if (val <= .05) else 'Not Significant'
    print(f'P-val: {val}  >>> {sig} <<<')
    print('\n')
    
    
# Run chi2 for a. Amphibian/Mammal, b. Fish/Mammal, c. Bird/Mammal, d. Bird/Non Vascular, e. Reptile/Vascular Plant 

run_chi2(0,3)
run_chi2(2,3)
run_chi2(1,3)
run_chi2(1,4)
run_chi2(5,6)


# From the results below, the differences between types of species and their conservation status are Significant
# comparing Fish to Mammal, Bird to Non Vascular Plants, and Reptile to Non Vascular Plant
```

```python
# Run everything, since there is a function to do it...

run_chi2(0,1)  # not
run_chi2(0,2)  # not
run_chi2(0,3)  # not
run_chi2(0,4)
run_chi2(0,5)  # not
run_chi2(0,6)
run_chi2(1,2)
run_chi2(1,3)  # not
run_chi2(1,4)
run_chi2(1,5)  # not
run_chi2(1,6)
run_chi2(2,3)
run_chi2(2,4)
run_chi2(2,5)  # not
run_chi2(2,6)
run_chi2(3,4)
run_chi2(3,5)
run_chi2(3,6)
run_chi2(4,5)
run_chi2(4,6)  # not
run_chi2(5,6)

# Results show that comparing any one to Vascular or Non Vascular Plants is Significant, except to each other.
```

### 4. Which animals were spotted most at each park? 

#### Codecademy methods

```python
# Comments: using the methods from codecademy...which I disagree with for several reasons:
#     a. The following is searching the species list and finding the most animals, which is not the same as finding the most 
#           animals spotted at each park. There could be twenty different types of elk each spotted once, or one type of 
#           elk spotted fifty times. The answer to the question 4. is the one type of elk spotted fifty times,  
#           not the twenty types spotted once. 
#     b. (major) The codecademy example did not clean 'observations' of duplicate data, and some other data on their output
#        can be seen to be counted twice.
#     c. (minor) The codecademy method is excluding all other categories which are on the observation list.

from itertools import chain
import string

def remove_punctuation(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text

species_animals = species_all[(species_all.category == 'Mammal')]
print(species_animals)
common_names = species_animals.common_names.apply(remove_punctuation).str.split().tolist()
common_names[:10]
```

```python
clean_rows = []

for item in common_names:
    item = list(dict.fromkeys(item))
    clean_rows.append(item)
    
clean_rows[:10]
```

```python
res = list(chain.from_iterable(i if isinstance(i, list) else [i] for i in clean_rows))
res[:10]
```

```python
words_counted = []

for i in res:
    x = res.count(i)
    words_counted.append((i,x))
    
pd.DataFrame(set(words_counted), columns = ['Word', 'Count']).sort_values('Count', ascending = False).head(20)
```

```python
# For this analysis, the term 'Bat' appears most. Create new column in species_animals that says whether the word 'bat' is
# in the 'common_names' column

species_all['is_bat'] = species_all.common_names.str.contains(r"\bBat\b", regex=True)
species_all.head(20)
```
